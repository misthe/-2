```python
import numpy as np
from matplotlib import pyplot as plt
import scienceplots
import pandas as pd
```

```python
plt.style.use(['science','notebook'])
```

```python
df=pd.read_csv('delay.csv',sep=';')
```

```python
plt.figure(figsize=(8,3))
plt.plot(df['tD'],df['N12'],'--',lw=0.9)
plt.plot(df['tD'],df['N12'],'o',color='red',ms=4)
plt.grid(False)
plt.xlabel('Time[ns]')
plt.ylabel('Coincidence counts $N_{12}$')
plt.title('Delay Curve')
plt.fill_between(df['tD'],7,13,alpha=0.3)
x=np.sqrt(df['N12'])
plt.errorbar(df['tD'],df['N12'],yerr=x,fmt='o',color='red',ms=4,ecolor='black',elinewidth=1.1,capsize=2)
```
